## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Quick inspection of missing values
print("Missing values before cleaning:")
print(titles_df.isnull().sum())

# Cleaning up data types and filling crucial structural missing fields
titles_df['imdb_score'] = titles_df['imdb_score'].fillna(titles_df['imdb_score'].median())
titles_df['imdb_votes'] = titles_df['imdb_votes'].fillna(0)
titles_df['tmdb_popularity'] = titles_df['tmdb_popularity'].fillna(titles_df['tmdb_popularity'].median())
titles_df['tmdb_score'] = titles_df['tmdb_score'].fillna(titles_df['tmdb_score'].median())
titles_df['age_certification'] = titles_df['age_certification'].fillna('Unrated')

# Helper function to safely parse string representations of lists (genres/countries)
def parse_list_column(val):
    if pd.isna(val):
        return []
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return [item.strip() for item in str(val).replace('[', '').replace(']', '').replace("'", "").split(',')]

titles_df['genres_clean'] = titles_df['genres'].apply(parse_list_column)
titles_df['countries_clean'] = titles_df['production_countries'].apply(parse_list_column)

print("\nData cleaning complete! New columns added: 'genres_clean' and 'countries_clean'.")
````

### What all manipulations have you done and insights you found?

Data Wrangling Manipulations & Insights:
During the data wrangling phase, we addressed missing values and structural formatting issues to prepare the dataset for seamless EDA:

Data Imputation: We filled missing numerical ratings (imdb_score, tmdb_score, tmdb_popularity) with their respective medians to prevent dropping valuable rows and skewing our distributions. Missing imdb_votes were filled with 0.

Categorical Handling: Missing values in the age_certification column were explicitly labeled as 'Unrated' so they could be tracked as a distinct category.

String Parsing (List Extraction): The genres and production_countries columns initially contained string representations of lists (e.g., "['comedy', 'drama']"). We utilized ast.literal_eval to parse these strings into actual, iterable Python lists, storing them in new columns (genres_clean and countries_clean). This manipulation is critical for allowing us to "explode" these columns later to analyze individual genre and country performance.
